## 7. Clean: `Clicks`
- Count nulls and check % missing
- Remove any non-numeric characters with regex
- Fill nulls with column mean
- Convert to `int` after rounding (clicks can't be fractional)

```python
len(df[df['Clicks'].isnull()])
```

```python
# Check for rows with decimal values (clicks shouldn't have decimals)
len(df[df['Clicks'].astype('str').str.contains(r'\d.\d',regex=True)])
```

```python
#df[df['Clicks'].astype('str').str.contains(r'\d\.\d|nan',case=False,regex=True)==False]['Clicks'].sample(20)
```

```python
len(df[df['Clicks'].isnull()])/len(df['Clicks'])*100
```

```python
df['Clicks'].skew()
```

```python
df['Clicks'] = df['Clicks'].replace(r'\D','')
df['Clicks'] = df['Clicks'].replace('', np.nan)
df['Clicks'] = df['Clicks'].fillna(df['Clicks'].mean())
df['Clicks'] = df['Clicks'].round().astype('int')                                 
df[df['Clicks'].isnull()]
```

## 9. Clean: `Cost`

- Check unique values for anomalies

## 13. Feature Engineering: `Conversion Rate`
Derive a new business metric: **Conversion Rate = (Conversions / Clicks) × 100**
This tells us what % of ad clicks resulted in a conversion.

```python
#Conversion_Rate = Conversions/Clicks×100
```

```python
df['Conversion Rate'] = np.where(df['Clicks'] > 0,(df['Conversions']/df['Clicks'])*100,0)
df['Conversion Rate'] = df['Conversion Rate'].round(2)
```

```python
df[['Clicks', 'Conversions', 'Conversion Rate']].sample(5)
```

## 14. Clean: `Sale_Amount`
- Strip `$` symbol and convert to `float`
- Fill nulls with mean
- Round to 2 decimal places (it's a monetary value)

```python
# Checking that don't match expected $XX format
df[~df['Sale_Amount'].str.contains(r'\$\d+',na=True)]
```

```python
df['Sale_Amount'] = df['Sale_Amount'].str.replace('$', '', regex=False)
```

```python
df[(df['Sale_Amount'].str.contains(r'\d+\.\d+'))&(df['Sale_Amount'].isna())]
len(df[df['Sale_Amount'].isnull()])/len(df['Sale_Amount'])*100
```

```python
df['Sale_Amount'] = df['Sale_Amount'].astype('float')
df['Sale_Amount'] = df['Sale_Amount'].fillna(df['Sale_Amount'].mean())
df['Sale_Amount'] = df['Sale_Amount'].round(2)
```

## 15. Clean: `Ad_Date`
- Dates come in mixed formats (DD-MM-YYYY, YYYY-MM-DD, etc.)
- Strip whitespace, parse with `format='mixed'` and `dayfirst=True`
- Standardize to `%d-%b-%y` format, then cast back to `datetime64`


```python
df[(df['Ad_Date'].str.contains(r'\d{2}-|/\d{2}-|/\d{4}'))|(df['Ad_Date'].str.contains(r'\d{4}-|/\d{2}-|/\d{2}'))].sample(20)
df[(~df['Ad_Date'].str.contains(r'\d{2}-|/\d{2}-|/\d{4}'))&(~df['Ad_Date'].str.contains(r'\d{4}-|/\d{2}-|/\d{2}'))]
```

```python
df['Ad_Date'] = df['Ad_Date'].str.strip()
```

```python
df['Ad_Date']=pd.to_datetime(df['Ad_Date'],errors='coerce',format='mixed',dayfirst=True)
```

```python
df['Ad_Date'] = df['Ad_Date'].dt.strftime('%d-%b-%y')
df['Ad_Date'] = df['Ad_Date'].astype('datetime64[s]')
```

```python
df['Ad_Date'].sample(5)
```

## 17. Clean: `Device`
- Capitalize and strip whitespace for consistency
- Encode as `category` dtype (low-cardinality column)

```python
df['Device'] = df['Device'].str.capitalize()
df['Device'] = df['Device'].str.strip()
```

```python
df['Device'].unique()
```

```python
df['Device'] = df['Device'].astype('category')
```

## 18. Clean: `Keyword`
- Title-case and strip whitespace
- Fix typos: `'Analitics'` and `'Analytic'` → `'Analytics'`
- Encode as `category`

```python
df['Keyword'] = df['Keyword'].str.title()
df['Keyword'] = df['Keyword'].str.strip()
```

```python
df.isnull().sum()
df['Keyword'].unique()
```

```python
df['Keyword'] = df['Keyword'].str.replace('Analitics','Analytics').str.replace('Analytic','Analytics')
```

## 19. Final Data Type Corrections
- `Ad_ID` → `object` (it's an identifier, not a number to compute on)
- `Location` → `object` (plain string column)

```python
df['Keyword']=df['Keyword'].astype('category')
df['Ad_ID'] = df['Ad_ID'].astype('object')
df['Location'] = df['Location'].astype('object')
```

## 20. Sort by Date
Sort the dataset chronologically by `Ad_Date` for better readability and time-series analysis downstream.

```python
df=df.sort_values(by=['Ad_Date'],ascending=True).reset_index(drop=True)
```
